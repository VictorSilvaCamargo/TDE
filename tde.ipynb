Escreva uma função em Python que compute o valor da função abaixo, emitindo uma
mensagem de erro caso a função receba um valor não permitido. Além disso, a função deverá
plotar um gráfico da função no intervalo entre ±10, destacando neste gráfico os pontos
relacionados ao valor de passado para a função.
$f(x) = \dfrac{1}{\sqrt{2-x}}$

```python
import matplotlib.pyplot as plt
import numpy as np

def funcao(x):
    if(x < 2):
        resultado = 1 / np.sqrt(2-x)
    else:
        print("Erro: o valor de x não pode ser",x,"pois o denominador não pode ser igual a 0 ou a raiz não pode ser negativa")
        return None
    
    x_values = np.linspace(-10, 10, 1000)
    y_values = 1 / (np.sqrt(2) - x_values)
    
    plt.plot(x_values, y_values)
    plt.scatter(x, resultado, color='red')
    plt.show()
    
    return resultado

funcao(1)
```

Usando a linguagem Python, sem utilizar as bibliotecas Sympy e Numpy, escreva uma função
que receba um valor 𝑛 no domínio dos números reais e devolva o limite da função dada a
seguir quando 𝑥 tende a 𝑛. $f(x) = \frac{3x^2 - 1}{\sqrt{x}}$






```python
def limite_f(n):
    h = 1e-10
    x = n + h
    fx = (3 * n ** 2 - 1) / (n ** 0.5)

    fxh = (3 * x ** 2 - 1) / (x ** 0.5)

    while abs(fxh - fx) > 1e-6:
        h /= 2
        x = n + h
        fx = fxh
        fxh = (3 * x ** 2 - 1) / (x ** 0.5)

    return fx

limite = limite_f(2)

print(limite)
```

Utilizando a linguagem Python, nenhuma biblioteca extra e List Comprehension crie uma
ff(x) = 4 (1 - $\sum_{1}^n \frac{1}{1+2n})$

```python
def f(n):
    somatoria = sum([1 / (1 + 2 * i) for i in range(1, n + 1)])
    resultado = 1 - somatoria
    return 4 * resultado

resultado = f(4)
print("Resultado da função f(n):", resultado)
```

Para encontrar o número de relógios que a ClockCom precisa vender para maximizar seu lucro, precisamos encontrar o valor de x que maximiza a função de lucro $f(x) = 300000 - 36000 + 75x^2 - \frac{x^3}{3}$. Podemos fazer isso encontrando a primeira derivada da função e igualando-a a zero para encontrar os pontos críticos.

A primeira derivada de $f(x)$ é $f'(x) = 150x - x^2$. Igualando $f'(x)$ a zero, temos:

\begin{align*}
150x - x^2 &= 0 \\
x(150 - x) &= 0
\end{align*}

Isso nos dá dois valores possíveis para x: 0 e 150. Para determinar qual desses valores maximiza o lucro, podemos calcular a segunda derivada de $f(x)$ e avaliá-la nesses pontos.

A segunda derivada de $f(x)$ é $f''(x) = 150 - 2x$. Avaliando $f''(x)$ em $x = 0$ e $x = 150$, temos:

\begin{align*}
f''(0) &= 150 > 0 \\
f''(150) &= -150 < 0
\end{align*}

Isso nos mostra que o lucro é maximizado quando $x = 150$. Portanto, a ClockCom precisa vender *150 relógios por mês* para maximizar seu lucro.

```python
import matplotlib.pyplot as plt
import numpy as np

def calcular_lucro(relogios_vendidos):
    lucro = 300000 - 36000 + 75 * relogios_vendidos*2 - (relogios_vendidos*3 / 3)
    return lucro

# Criar um array de valores de x para o gráfico
x = np.linspace(0, 200, 400)

# Calcular os valores de y (lucro) correspondentes
y = calcular_lucro(x)

# Plotar o gráfico
plt.plot(x, y, label="Função de Lucro")
plt.scatter(150, calcular_lucro(150), color='r', label='Lucro Máximo em x=150')
relogios_vendidos = 100 #quantidade de relogios vendidos no mes para o grafico
plt.scatter(relogios_vendidos, calcular_lucro(relogios_vendidos), color='g', label=f'Quantidade Vendida em x={relogios_vendidos}')
plt.xlabel('Quantidade de Relógios Vendidos')
plt.ylabel('Lucro')
plt.legend()
plt.title('Função de Lucro da ClockCom')
plt.grid(True)
plt.show()
```
